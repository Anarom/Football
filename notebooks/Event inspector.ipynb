```python
skipped_quals = []
def get_set(inspected_type, log=False):
    generator = event_generator(count = 500,mandatory_types={inspected_type})
    new_quals = set()
    for event in generator:
        quals = event['qualifiers']
        for qual in quals:
            if qual['type']['value'] in skipped_quals:
                continue
            elif qual['type']['displayName'] not in new_quals:
                new_quals.add(qual['type']['displayName'])
                if log:
                    print(qual)
    return new_quals
```

```python
db = database.DataBase()
sql = 'SELECT typeId, typeText from event_types'
db.crs.execute(sql)
types = db.crs.fetchall()
db.close()
```

```python
d = {}
for code,inspected_type in types:
    d[inspected_type] = get_set(inspected_type)
```
