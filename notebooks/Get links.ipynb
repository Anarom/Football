```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import re

options = webdriver.ChromeOptions()
options.add_argument('--log-level=3')
options.add_argument('--incognito')
options.add_argument('--no-first-run')
options.add_argument('--no-service-autorun')
button = '.button.ui-state-default.rc-l'
regexp = re.compile('https://1xbet\.whoscored\.com/Matches\/(?P<outerID>\d*)\/Live\/(?P<country>[A-Z][a-z]*)-(?P<league>.*?)-(?:(?P<startYear>\d{4})(?:-(?P<endYear>\d{4}))?)-(?P<teams>.*)')
```

```python
driver = webdriver.Chrome('chromedriver.exe', options=options)

ids = {}
with open ('competitions.txt', 'r', encoding = 'utf-8') as f_imp:
    for line in f_imp:
        ids[line] = set()
        driver.get(line)
        try:
            popup = driver.find_element(By.CSS_SELECTOR, 'body > div.webpush-swal2-container.webpush-swal2-center.webpush-swal2-fade.webpush-swal2-shown > div > div.webpush-swal2-header > button')
            popup.click()
        except NoSuchElementException:
            pass
        
        while True:
            elements = driver.find_elements(By.CSS_SELECTOR, '.result-1')
            links = set([elem.get_attribute('href') for elem in elements])
            for link in links:
                regexp_result = re.search(regexp, link)
                print(regexp_result['outerID'])
                ids[line].add(regexp_result['outerID'])
            title = driver.find_element(By.CSS_SELECTOR, button).get_attribute('title')
            if title == 'No data for previous week':
                break
            else:
                driver.find_element(By.CSS_SELECTOR, button).click()
                sleep(1)
driver.quit()
```

```python
from sqlalchemy import create_engine, update
from sqlalchemy.ext.automap import automap_base
import datetime
from sqlalchemy.orm import Session
engine = create_engine('postgresql://develop:}c%4Z>n~M<3p:Em\@localhost:5432/football')
Base = automap_base()
Base.prepare(autoload_with=engine, schema='core')
Tournament = Base.classes.tournaments
Game = Base.classes.games
```

```python
driver = webdriver.Chrome('chromedriver.exe', options=options)
regexp = re.compile('https://1xbet\.whoscored\.com\/Regions\/(?P<regionID>\d*)\/Tournaments\/(?P<tournamentID>\d*)\/Seasons\/(?P<seasonID>\d*)')
with Session(engine) as session:
    
    for k, v in ids.items():
        link = k.rstrip()
        season_id = re.search(regexp, link)['seasonID']
        driver.get(k)
        try:
            popup = driver.find_element(By.CSS_SELECTOR, 'body > div.webpush-swal2-container.webpush-swal2-center.webpush-swal2-fade.webpush-swal2-shown > div > div.webpush-swal2-header > button')
            popup.click()
        except NoSuchElementException:
            pass
        data = driver.execute_script('return tables')[0]
        title = driver.find_element(By.CSS_SELECTOR, 'option[selected="selected"]').text
        country = driver.find_element(By.CSS_SELECTOR, '#breadcrumb-nav > span.iconize.iconize-icon-left').text
        t = Tournament()
        t.id = season_id
        t.name = title
        t.country = country
        start_date = [int(x) for x in data['startDate'].split('-')]
        end_date = [int(x) for x in data['endDate'].split('-')]
        t.start_date = datetime.date(start_date[2],start_date[1],start_date[0])
        t.end_date = datetime.date(end_date[2],end_date[1],end_date[0])
        t.type = 1
        session.add(t)
        for game_id in v:
            session.execute(update(Game).where(Game.id == game_id).values(tournament=t.id))
        session.commit()
driver.quit()
```
