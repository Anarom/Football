```python
import config
import os
import json
from sqlalchemy import create_engine, select
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

```python
filenames = [filename for filename in os.listdir(os.path.join(os.getcwd(),'data'))]
engine = create_engine('postgresql://develop:}c%4Z>n~M<3p:Em\@localhost:5432/football')
Base = automap_base()
Base.prepare(autoload_with=engine, schema='core')
Team = Base.classes.teams
Game = Base.classes.games
```

```python
for filename in filenames:
    game_records = []
    with open(os.path.join(os.getcwd(),'data',filename), 'r', encoding='utf-8') as file:
        for line in file:
            game_records.append(json.loads(line)['config']['params']['args'])
    with Session(engine) as session:
        for record in game_records:
            if not record.get('matchCentreData'):
                continue
            if not record['matchCentreData'].get('home'):
                continue
            ext_data = record['matchCentreData']
            game = Game()
            game.id  = record['matchId']
            game.attendance = ext_data.get('attendance')
            game.venue = ext_data.get('venueName')
            game.start_time = ext_data.get('startTime')
            game.score = ext_data.get('score')
            game.period_end_minutes = ext_data.get('periodEndMinutes')
            game.weather = int(ext_data['weatherCode'] if ext_data.get('weatherCode') else -1)
            expanded_max_minute = ext_data.get('expanded_max_minute')
            if expanded_max_minute and expanded_max_minute < 200:
                game.expanded_max_minute = expanded_max_minute
            minute_expanded = ext_data.get('minuteExpanded')
            if minute_expanded and minute_expanded < 200:
                game.minute_expanded = minute_expanded
            referee = ext_data.get('referee')
            game.referee = referee['officialId'] if referee else None
            query_result = session.execute(select(Game).where(Game.id == game.id))
            game.home_team = ext_data['home']['teamId']
            game.away_team = ext_data['away']['teamId']
            if not query_result.first():    
                session.add(game)
            
            for team in ('home', 'away'):
                new_team = Team()
                new_team.id = ext_data[team]['teamId'] 
                new_team.name = ext_data[team]['name']
                new_team.country = ext_data[team]['countryName']
                #new_team.manager = ext_data[team]['managerName']
                query_result = session.execute(select(Team).where(Team.id == new_team.id))
                if not query_result.first():
                    session.add(new_team)

        session.commit()
    print(f'{filename} done')
```
